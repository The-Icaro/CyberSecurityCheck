```python
import json
import requests
from fuzzywuzzy import process
import pandas as pd
import random
```

```python
namesData = requests.get('https://cspinheiro.github.io/interpol.json')
testData = pd.read_csv('CheckPointData - Data.csv')

cleanData = testData['Frases'].dropna()
```

```python
def handleBlockedPersons(name, th):

    blockedNamesData = namesData.json()['interpol']

    blocked = []

    for data in blockedNamesData:
      for i, v in data.items():
        blocked.append(v)

    searchList = process.extract(str(name), blocked)

    for i in searchList:
      if i[1] >= th:
        return 'Proibido'

      return 'Permitido'
```

```python
th = 60
while th <= 95:
  print(f'TH -> {th}')
  print('-' * 10)
  for i in cleanData:
    blocked = handleBlockedPersons(i, th)
    print(f'{i} -> {blocked}')
  print('')
  th += 5
```
